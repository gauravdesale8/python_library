Linear Plot

```python
import matplotlib.pyplot as plt

x = [2,3,5,7,8]
y = [3,5,2,9,6]

plt.plot(x,y)

plt.show()
```

Bar plot

```python
import matplotlib.pyplot as plt

x = [23,55,65,56,34]
y = [45,69,32,15,28]

plt.bar(x,y)

plt.show()
```

Colors in graph

```python
import matplotlib.pyplot as plt

x = [1,2,3,4]
y = [2,4,6,8]
z = ['r','g','b','y']

plt.bar(x,y, color = z)

plt.show()
```
