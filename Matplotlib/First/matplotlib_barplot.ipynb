Barplot with Matplotlib

```python
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,75,70,85]

plt.bar(x,y)
plt.show()
```

```python
# adding labels to the plot

import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,75,65,83]

plt.xlabel("Programming Languages")
plt.ylabel("Percentage")
plt.title("Languages")

plt.bar(x,y)
plt.show()
```

```python
# adding fontsize
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,80]

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

plt.bar(x,y)
plt.show()
```

```python
# width size
# adding fontsize
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,80]

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

plt.bar(x,y, width=0.4)
plt.show()
```

```python
# adding different color# adding fontsize
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,80]

plt.xlabel("Programming", fontsize = 13)
plt.ylabel("Percentage", fontsize = 13)
plt.title("Languages", fontsize = 13)

plt.bar(x,y, width = 0.4, color = 'y')
plt.show()
```

```python
# adding different different colors
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,45,63,76]
z = ['y','m','g','r']

plt.xlabel('Programming', fontsize = 12.5)
plt.ylabel('Percentage', fontsize = 12.5)
plt.title('languages', fontsize = 12.5)

plt.bar(x,y,width = 0.4, color = z)
plt.show()
```

```python
# align edge parameter
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,86]
c = ['r','y','c','b']

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

plt.bar(x,y,width=0.4,color=c,align='edge')
plt.show()
```

```python
# two bar plots
import matplotlib.pyplot as plt

x = ['Python','C','C++','Java']
z = [55,65,35,56]
y = [90,86,80,70]

plt.xlabel('Programming')
plt.ylabel('Percentage')
plt.title('Languages')

plt.bar(x,y,width=0.4,color='y',label='Yellow')
plt.bar(x,z,width=0.4,color='c',label='Cyan')
plt.legend()
plt.show()
```

```python
# width
import matplotlib.pyplot as plt

x = ['Python','C','C++','Java']
y = [90,86,80,70]
z = [55,65,35,56]

plt.xlabel('Programming',fontsize=10)
plt.ylabel('Languages',fontsize=10)
plt.title('Percentage',fontsize=10)

width = 0.2
plt.bar(x,y,width,color='g',label='green')
plt.bar(x,z,width,color='b',label='blue')

plt.legend()
plt.show()

```

```python
# creating numpy array instead of list
import matplotlib.pyplot as plt
import numpy as np

x = ['Python','C','C++','Java']
y = [90,86,80,70]
z = [55,65,35,56]

p = np.arange(len(x))
plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

width=0.2
plt.bar(p,y,width,color='r',label='red')
plt.bar(p,z,width,color='g',label='green')
plt.legend()
plt.show()

```

```python
# side by side bar graph
import matplotlib.pyplot as plt
import numpy as np

x = ['Python','C','C++','Java']
y = [90,86,80,70]
z = [55,65,35,56]

width=0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)


plt.bar(p,y,width,color='r',label='red')
plt.bar(p1,z,width,color='g',label='green')

plt.legend()
plt.show()
```

```python
# adding proper labels
import matplotlib.pyplot as plt
import numpy as np

x = ['Python','C','C++','Java']
y = [90,86,80,70]
z = [55,65,35,56]

width=0.2
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)


plt.bar(p,y,width,color='r',label='red')
plt.bar(p1,z,width,color='g',label='green')

plt.xticks(p+width,x)
plt.legend()
plt.show()
```

```python
# aligning labels in centre
import matplotlib.pyplot as plt
import numpy as np

x = ['Python','C','C++','Java']
y = [90,86,80,70]
z = [55,65,35,56]

width=0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)


plt.bar(p,y,width,color='r',label='red')
plt.bar(p1,z,width,color='g',label='green')

plt.xticks(p + width/2, x)
plt.legend()
plt.show()
```

```python
# adding rotation to labels
import matplotlib.pyplot as plt
import numpy as np

x = ['Python','C','C++','Java']
y = [90,86,80,70]
z = [55,65,35,56]

width=0.2
p = np.arange(len(x))
p1 = [j + width for j in p]

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)


plt.bar(p,y,width,color='r',label='red')
plt.bar(p1,z,width,color='g',label='green')

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
# horizontal bar graph
import matplotlib.pyplot as plt


x = ['Python','C','C++','Java']
y = [90,86,80,70]
z = [55,65,35,56]



plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)


plt.barh(x,y,color='r',label='red')
plt.bar(x,z,color='g',label='green')


plt.legend()
plt.show()
```
