Pie Plot

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]

plt.pie(x)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ["c","c++","java","python"]

plt.pie(x,labels=y)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
x1 = [20,30,40,50]
y = ['c','c++','java','python']


plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,labels=y,radius=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
x1 = [20,30,40,50]
y = ['c','c++','java','python']
c = ["r","g","b","y"]


plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=1,colors=c)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ['c','c++','java','python']

plt.pie(x,labels=y,radius=1.5)
plt.pie([1],colors="w")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ['c','c++','java','python']


plt.pie(x,labels=y,radius=1.5)
cr = plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)
plt.show()
```
