Stem Plot

```python
import matplotlib.pyplot as plt

x = [2,4,1,3,5,6]
y = [3,5,2,1,6,3]

plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,6,6)
y = np.random.randint(1,6,6)

plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,6,6)
y = np.random.randint(1,6,6)

plt.stem(x,y,linefmt=":")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,6,6)
y = np.random.randint(1,6,6)

plt.stem(x,y,linefmt=":",markerfmt="r+")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,6,6)
y = np.random.randint(1,6,6)

plt.stem(x,y,basefmt="g",markerfmt="r")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,6,6)
y = np.random.randint(1,6,6)

plt.stem(x,y,basefmt="g",markerfmt="r",label="Python")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,6,6)
y = np.random.randint(1,6,6)

plt.stem(x,y,basefmt="g",linefmt=":",markerfmt="r^",use_line_collection=False,label="python")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,6,6)
y = np.random.randint(1,6,6)

plt.stem(x,y,linefmt=":",markerfmt="r>",basefmt="g",use_line_collection=False,orientation="horizontal",label="python")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,6,6)
y = np.random.randint(1,6,6)

plt.stem(x,y,basefmt="g",linefmt="--",markerfmt="r^",use_line_collection=False,label="python")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1,6,6)
y = np.random.randint(1,6,6)

plt.stem(x,y,basefmt="g",linefmt="-.",markerfmt="r^",use_line_collection=False,label="python")
plt.legend()
plt.show()
```
