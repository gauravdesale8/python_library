```python
import matplotlib.pyplot as plt

# Data for area plot
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 6, 8, 9]
y2 = [2, 2, 7, 10, 12]
y3 = [1, 3, 5, 7, 9]

# Data for stack plot
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [2, 4, 5, 1, 3]
values2 = [1, 3, 2, 6, 4]
values3 = [3, 1, 4, 2, 5]

# Create area plot
plt.subplot(1, 2, 1)
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.plot(x, y3, label='Line 3')
plt.fill_between(x, y1, y2, alpha=0.3, color='blue')
plt.fill_between(x, y2, y3, alpha=0.3, color='green')
plt.fill_between(x, y3, [0]*len(x), alpha=0.3, color='orange')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Area Plot')
plt.legend()

# Create stack plot
plt.subplot(1, 2, 2)
plt.stackplot(categories, values1, values2, values3, labels=['Value 1', 'Value 2', 'Value 3'])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stack Plot')
plt.legend()

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
