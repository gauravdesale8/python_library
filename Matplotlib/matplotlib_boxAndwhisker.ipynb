Box plot

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,70,7)

plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,70,7)

plt.boxplot(x,notch=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,70,7)

plt.boxplot(x,vert=False)
plt.show()
```
