```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(0)
data = np.random.normal(0, 1, 100)

# Create a figure and axis
fig, ax = plt.subplots()

# Create the boxplot
boxplot = ax.boxplot(data, vert=False, patch_artist=True)

# Customize the boxplot colors
box_colors = ['#1f77b4']
whisker_colors = ['#757575']
boxplot['boxes'][0].set(facecolor=box_colors[0], linewidth=2)
boxplot['whiskers'][0].set(color=whisker_colors[0], linewidth=2)
boxplot['whiskers'][1].set(color=whisker_colors[0], linewidth=2)
boxplot['caps'][0].set(color=whisker_colors[0], linewidth=2)
boxplot['caps'][1].set(color=whisker_colors[0], linewidth=2)
boxplot['medians'][0].set(color=whisker_colors[0], linewidth=2)
boxplot['fliers'][0].set(markerfacecolor=whisker_colors[0], marker='o', markeredgecolor='black')

# Set labels and title
ax.set_yticklabels(['Data'])
ax.set_title('Whisker and Box Plot')

# Show the plot
plt.show()
```
