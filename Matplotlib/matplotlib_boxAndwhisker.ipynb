Box plot

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,70,7)

plt.boxplot(x)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,70,7)

plt.boxplot(x,notch=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,70,7)

plt.boxplot(x,vert=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,50,5)

plt.boxplot(x,widths=0.8)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,50,5)

plt.boxplot(x,labels=["Python"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,40,5)

plt.boxplot(x,patch_artist=True)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x,showmeans=True,labels=["python"])
plt.show()
```

```python
# 120 is outlier
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["python"],showmeans=True)
plt.show()
```

```python
# 120 is outlier
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["python"],showmeans=True,whis=3.5)
plt.show()
```

```python
# 120 is outlier
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,labels=["python"],showmeans=True,sym="b+")
plt.show()
```

```python

import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x,labels=["python"],showmeans=True,boxprops=dict(color="r"))
plt.show()
```

```python
# 120 is outlier
import matplotlib.pyplot as plt

x = [10,20,30,40,50,60,70]

plt.boxplot(x,labels=["python"],showmeans=True,boxprops=dict(color="r"),capprops=dict(color="b"))
plt.show()
```

```python

x = [10,20,30,40,50,60,70]
import matplotlib.pyplot as plt


plt.boxplot(x,labels=["python"],showmeans=True, boxprops=dict(color="r"),whiskerprops=dict(color="b"))
plt.show()
```

```python
# 120 is outlier
import matplotlib.pyplot as plt
import numpy as np

x = [10,20,30,40,50,60,70]
y = [10,20,50,60,80,90,55]

z = [y,x]
plt.boxplot(z,labels=["python"],showmeans=True)
plt.show()
```
