Scatter plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [3,5,2,6,5,8]

plt.scatter(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6]
no = [4,5,3,6,8,4]

plt.title("Scatter Plot",fontsize =15)
plt.xlabel("Days", fontsize = 12)
plt.ylabel("No.",fontsize = 12)

plt.scatter(day,no)
plt.show()
```

```python
# adding different colors
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [23,65,34,67,45,26,75]
c = ["r","g","y","b","r","g","y"]

plt.title("Scatter Plot")
plt.xlabel("Days")
plt.ylabel("No.")

plt.scatter(x,y,color = c)
plt.show()
```

```python
# changing the size of dots
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6]
no = [23,45,64,35,27,15]
colors = ["r","g","b","y","m","c"]
size = [120,345,500,430,200,100]

plt.title("Scatter plot")
plt.xlabel("Days")
plt.ylabel("No.")

plt.scatter(day,no,c=colors,s=size)
plt.show()
```

```python
# using cmap(colormap)
import matplotlib.pyplot as plt

day = [1,2,3,4,5,6]
no = [23,45,64,35,27,15]
colors = [12,34,56,43,53,23]
size = [120,345,500,430,200,100]

plt.title("Scatter plot")
plt.xlabel("Days")
plt.ylabel("No.")

plt.scatter(day,no,c=colors,s=size,edgecolors="k",linewidths=1,cmap="viridis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

day = [1,2,3,4,5,6]
no = np.random.randint(10,70,6)
colors = np.random.randint(1,99,6)
size = np.random.randint(100,600,6)

plt.title("Scatter plot")
plt.xlabel("Days")
plt.ylabel("No.")

plt.scatter(day,no,s=size,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

day = [1,2,3,4,5,6]
no = np.random.randint(10,70,6)
colors = np.random.randint(1,99,6)
size = np.random.randint(100,600,6)

plt.title("Scatter plot")
plt.xlabel("Days")
plt.ylabel("No.")

plt.scatter(day,no,s=size,c=colors,cmap='viridis')
t = plt.colorbar()
t.set_label("ColorBar")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

day = [1,2,3,4,5,6]
no = np.random.randint(10,70,6)
colors = np.random.randint(1,99,6)
size = np.random.randint(100,600,6)

plt.title("Scatter plot")
plt.xlabel("Days")
plt.ylabel("No.")

plt.scatter(day,no,s=size,c=colors,cmap='viridis')
t = plt.colorbar()
t.set_label("ColorBar",fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

day = [1,2,3,4,5,6]
no = np.random.randint(10,70,6)
no2 = np.random.randint(10,70,6)
colors = np.random.randint(1,99,6)
size = np.random.randint(100,600,6)

plt.title("Scatter plot")
plt.xlabel("Days")
plt.ylabel("No.")

plt.scatter(day,no,s=size,c=colors,cmap='cubehelix')
plt.scatter(day,no2,color='r',s=size)
t = plt.colorbar()
t.set_label("ColorBar")
plt.show()
```
