Barplot with Matplotlib

```python
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,75,70,85]

plt.bar(x,y)
plt.show()
```

```python
# adding labels to the plot

import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,75,65,83]

plt.xlabel("Programming Languages")
plt.ylabel("Percentage")
plt.title("Languages")

plt.bar(x,y)
plt.show()
```

```python
# adding fontsize
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,80]

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

plt.bar(x,y)
plt.show()
```

```python
# width size
# adding fontsize
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,80]

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

plt.bar(x,y, width=0.4)
plt.show()
```

```python
# adding different color# adding fontsize
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,80]

plt.xlabel("Programming", fontsize = 13)
plt.ylabel("Percentage", fontsize = 13)
plt.title("Languages", fontsize = 13)

plt.bar(x,y, width = 0.4, color = 'y')
plt.show()
```

```python
# adding different different colors
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,45,63,76]
z = ['y','m','g','r']

plt.xlabel('Programming', fontsize = 12.5)
plt.ylabel('Percentage', fontsize = 12.5)
plt.title('languages', fontsize = 12.5)

plt.bar(x,y,width = 0.4, color = z)
plt.show()
```

```python
# align edge parameter
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,86]
c = ['r','y','c','b']

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

plt.bar(x,y,width=0.4,color=c,align='edge')
plt.show()
```

```python
# edge color and line width
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,45]
c = ['r','b','y','c']

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

plt.bar(x,y,color=c,width=0.4,edgecolor = 'g',linewidth = 12)
plt.show()
```

```python
# linestyle
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,45]
c = ['r','b','y','c']

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

plt.bar(x,y,color=c,width=0.4,edgecolor = 'g',linewidth = 8,linestyle=':')
plt.show()
```

```python
# aplha para for blur
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,45]
c = ['r','b','y','c']

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

plt.bar(x,y,color=c,width=0.4,edgecolor = 'g',linewidth = 7, linestyle=':',alpha=0.5)
plt.show()
```

```python
# adding labels
import matplotlib.pyplot as plt

x = ['python','c','c++','java']
y = [90,65,78,45]
c = ['r','b','y','c']

plt.xlabel("Programming", fontsize = 15)
plt.ylabel("Percentage", fontsize = 15)
plt.title("Languages", fontsize = 15)

plt.bar(x,y,color=c,width=0.4,edgecolor = 'g',label = 'red')
plt.legend()
plt.show()
```
