```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, -3, 1, -4, 3]

# Create the stem plot
plt.stem(x, y, linefmt='b-', markerfmt='bo', basefmt='r-')

# Customize the plot
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stem Plot')
plt.grid(True)

# Display the plot
plt.show()
```
